## **Exercício** 10: Calcule a quantidade total de atribuições do algoritmo definido pela função ```metodo_da_bisseccao``` abaixo.

````python
def f(x):
  result = x**2 - 4*x + 3   #1 at
  return result

def metodo_da_bisseccao(N):

      a = 0         #1 at

      b = 2         #1at

      fa = f(a)     #2at
      fb = f(b)     #2at
      comp_ab = fa*fb   #1at
      if comp_ab >= 0:
          print("Bisection method fails.")
          return None

      a_n = a     #1 at
      b_n = b     #1 at
      for n in range(1,N+1):    #roda em n vezes   ao todo são 10*n atribuições nesse laço

          m_n = (a_n + b_n)/2   #1
          f_m_n = f(m_n)        #2

          fan = f(a_n)          #2
          fbn = f(b_n)          #2

          comp = fan*f_m_n      #1
          if comp < 0:          #2
              a_n = a_n
              b_n = m_n
          elif fbn*f_m_n < 0:   #2 mesma quantidade de atribuições do anterior, então só conto uma vez.
              a_n = m_n
              b_n = b_n
          elif f_m_n == 0:
              print("Solução exata encontrada.")
              return m_n
          else:
              print("Método da Bissecção falhou.")
              return None

      result = (a_n + b_n)/2    #1 at

      return result

#T_metodo_da_bisseccao(n) = 10n+11

metodo_da_bisseccao(5)
````
