###**Exercício 1**: Escreva um código em Python que verifica se duas árvores binárias são iguais.

**Obs**.: (1) Uma árvore binária $T_1$ é igual a outra árvore binária $T_2$ se, e somente se, todo vértice de uma apresenta os mesmos parentes na outra.

(2) O código pode levar em conta tanto um objeto da classe `Arvore_Binaria` tratada em aula quanto a sua notação em parêntesis / colchêtes aninhados.

```python
# Solução:~>

def iguais(x,y):
  if x is None and y is None:

    return True

  if x is not None and y is not None:
    return ((x.j == y.j) and iguais(x.direita, y.direita) and iguais(x.esquerda, y.esquerda))

  return False
```

```python
class Arvore_Binaria_de_Busca:

  __raiz = None
  __tamanho = 0

  def __init__(self, raiz):
    self.set_raiz(raiz)
    if raiz is not None:
      self.add_tamanho()

  def insere_vertice(self, novo_vertice):

      if self.get_raiz() is None:
        self.set_raiz(novo_vertice)

      else:
        self.aux_insere(novo_vertice, self.get_raiz())

  def aux_insere(self, novo_vertice, possivel_pai):
    chave = novo_vertice.get_chave()

    if possivel_pai.get_chave() < chave:
      filho_dir = possivel_pai.get_filho_direita()

      if filho_dir is not None:
        self.aux_insere(novo_vertice, filho_dir)

      else:
        possivel_pai.set_filho_direita(novo_vertice)
        novo_vertice.set_pai(possivel_pai)

    elif possivel_pai.get_chave() > chave:
      filho_esq = possivel_pai.get_filho_esquerda()

      if filho_esq is not None:
        self.aux_insere(novo_vertice, filho_esq)

      else:
        possivel_pai.set_filho_esquerda(novo_vertice)
        novo_vertice.set_pai(possivel_pai)

    else:
      print("Chave já existente na árvore!")

  def busca(self, chave):

      temp = self.get_raiz()

      while temp.get_chave() != chave and temp is not None:
        chave_temporaria = temp.get_chave()

        if chave < chave_temporaria:
          temp = temp.get_filho_esquerda()

        elif chave > chave_temporaria:
          temp = temp.get_filho_direita()

        else:

          break

      if temp is None:
        print("Essa chave não existe na árvore!")

      else:

        return temp

  def remove_vertice(self, chave_a_ser_removida):

      temp = self.get_raiz()

      while temp.get_chave() != chave_a_ser_removida and temp is not None:

        chave_temporaria = temp.get_chave()

        if chave_a_ser_removida < chave_temporaria:
          temp = temp.get_filho_esquerda()

        elif chave_a_ser_removida > chave_temporaria:
          temp = temp.get_filho_direita()

        else:

          break

      if temp is None:
        print("Essa chave já não existe na árvore!")

      else:
        self.analisa_casos_de_remocao(temp)
        self.sub_tamanho()

  def analisa_casos_de_remocao(self, vertice_a_ser_removido):

      filho_esq = vertice_a_ser_removido.get_filho_esquerda()
      filho_dir = vertice_a_ser_removido.get_filho_direita()
      pai       = vertice_a_ser_removido.get_pai()

      if filho_esq is None and filho_dir is None:
        self.caso_sem_filhos(vertice_a_ser_removido, pai)

      elif filho_esq is not None and filho_dir is not None:
        self.caso_filho_duplo(vertice_a_ser_removido, filho_esq, filho_dir, pai)

      else:
        self.caso_filho_unico(vertice_a_ser_removido, filho_esq, filho_dir, pai)

  def caso_sem_filhos(self, vertice_a_ser_removido, pai):
      if pai is not None:

        if pai.get_filho_direita().get_chave() == vertice_a_ser_removido.get_chave():
          pai.set_filho_direita(None)

        else:
          pai.set_filho_esquerda(None)

      vertice_a_ser_removido = None

  def caso_filho_unico(self, vertice_a_ser_removido, filho_esq, filho_dir, pai):

    if filho_dir is not None:
      filho = filho_dir

    elif filho_esq is not None:
      filho = filho_esq

    else:
      raise KeyError("Ambos os filhos de um vértice com 1 filho são nulos.")

    filho.set_pai(pai)

    if pai is None:
      self.set_raiz(None)
      self.set_raiz(filho)

    else:

      if vertice_a_ser_removido.get_chave() == pai.get_filho_direita().get_chave():
        pai.set_filho_direita(filho)

      else:
        pai.set_filho_esquerda(filho)

  def detecta_sucessor(self, vertice):
    sucessor = vertice.get_filho_esquerda()
    filho_a_direita_do_possivel_sucessor = sucessor.get_filho_direita()

    while filho_a_direita_do_possivel_sucessor is not None:
      sucessor = filho_a_direita_do_possivel_sucessor
      filho_a_direita_do_possivel_sucessor = sucessor.get_filho_direita()

    return sucessor

  def caso_filho_duplo(self, vertice_a_ser_removido, filho_esq, filho_dir, pai):

    sucessor = self.detecta_sucessor(vertice_a_ser_removido)
    chave_sucessor = sucessor.get_chave()
    vertice_a_ser_removido.set_chave(chave_sucessor)

    if sucessor.get_filho_direita() is None and sucessor.get_filho_esquerda() is None:
      self.caso_sem_filhos(sucessor, sucessor.get_pai())

    else:
      self.caso_filho_unico(sucessor, sucessor.get_filho_esquerda(), sucessor.get_filho_direita(), sucessor.get_pai())

  def add_tamanho(self):
    self.__tamanho = self.__tamanho + 1

  def sub_tamanho(self):
    self.__tamanho = self.__tamanho - 1

  def percurso_pre_order(self, elementos_visitados = [], vertice = None):

    if vertice is None:
      vertice = self.get_raiz()
      elementos_visitados = []

    elementos_visitados.append(vertice.get_chave())

    if vertice.get_filho_esquerda() is not None:
      self.percurso_pre_order(elementos_visitados, vertice.get_filho_esquerda())

    if vertice.get_filho_direita() is not None:
      self.percurso_pre_order(elementos_visitados, vertice.get_filho_direita())

    return elementos_visitados

  def percurso_in_order(self, elementos_visitados = [], vertice = None):

    if vertice is None:
      vertice = self.get_raiz()
      elementos_visitados = []

    if vertice.get_filho_esquerda() is not None:
      self.percurso_in_order(elementos_visitados, vertice.get_filho_esquerda())

    elementos_visitados.append(vertice.get_chave())

    if vertice.get_filho_direita() is not None:
      self.percurso_in_order(elementos_visitados, vertice.get_filho_direita())

    return elementos_visitados

  def percurso_post_order(self, elementos_visitados = [], vertice = None):

    if vertice is None:
      vertice = self.get_raiz()
      elementos_visitados = []

    if vertice.get_filho_esquerda() is not None:
      self.percurso_post_order(elementos_visitados, vertice.get_filho_esquerda())

    if vertice.get_filho_direita() is not None:
      self.percurso_post_order(elementos_visitados, vertice.get_filho_direita())

    elementos_visitados.append(vertice.get_chave())

    return elementos_visitados

  def set_raiz(self, raiz):

    if self.__raiz is None or raiz is None:
      self.__raiz = raiz

    else:
      if raiz.get_chave() > self.__raiz.get_chave():
        raiz.set_filho_esquerda(self.__raiz)

      elif raiz.get_chave() < self.__raiz.get_chave():
        raiz.set_filho_direita(self.__raiz)

      self.__raiz.set_pai(raiz)

      self.__raiz = raiz

  def get_raiz(self):
    return self.__raiz
```

```python
class Arvore_Binaria_de_Busca:

  __raiz = None
  __tamanho = 0

  def __init__(self, raiz):
    self.set_raiz(raiz)
    if raiz is not None:
      self.add_tamanho()

  def insere_vertice(self, novo_vertice):

      if self.get_raiz() is None:
        self.set_raiz(novo_vertice)

      else:
        self.aux_insere(novo_vertice, self.get_raiz())

  def aux_insere(self, novo_vertice, possivel_pai):
    chave = novo_vertice.get_chave()

    if possivel_pai.get_chave() < chave:
      filho_dir = possivel_pai.get_filho_direita()

      if filho_dir is not None:
        self.aux_insere(novo_vertice, filho_dir)

      else:
        possivel_pai.set_filho_direita(novo_vertice)
        novo_vertice.set_pai(possivel_pai)

    elif possivel_pai.get_chave() > chave:
      filho_esq = possivel_pai.get_filho_esquerda()

      if filho_esq is not None:
        self.aux_insere(novo_vertice, filho_esq)

      else:
        possivel_pai.set_filho_esquerda(novo_vertice)
        novo_vertice.set_pai(possivel_pai)

    else:
      print("Chave já existente na árvore!")

  def busca(self, chave):

      temp = self.get_raiz()

      while temp.get_chave() != chave and temp is not None:
        chave_temporaria = temp.get_chave()

        if chave < chave_temporaria:
          temp = temp.get_filho_esquerda()

        elif chave > chave_temporaria:
          temp = temp.get_filho_direita()

        else:

          break

      if temp is None:
        print("Essa chave não existe na árvore!")

      else:

        return temp

  def remove_vertice(self, chave_a_ser_removida):

      temp = self.get_raiz()

      while temp.get_chave() != chave_a_ser_removida and temp is not None:

        chave_temporaria = temp.get_chave()

        if chave_a_ser_removida < chave_temporaria:
          temp = temp.get_filho_esquerda()

        elif chave_a_ser_removida > chave_temporaria:
          temp = temp.get_filho_direita()

        else:

          break

      if temp is None:
        print("Essa chave já não existe na árvore!")

      else:
        self.analisa_casos_de_remocao(temp)
        self.sub_tamanho()

  def analisa_casos_de_remocao(self, vertice_a_ser_removido):

      filho_esq = vertice_a_ser_removido.get_filho_esquerda()
      filho_dir = vertice_a_ser_removido.get_filho_direita()
      pai       = vertice_a_ser_removido.get_pai()

      if filho_esq is None and filho_dir is None:
        self.caso_sem_filhos(vertice_a_ser_removido, pai)

      elif filho_esq is not None and filho_dir is not None:
        self.caso_filho_duplo(vertice_a_ser_removido, filho_esq, filho_dir, pai)

      else:
        self.caso_filho_unico(vertice_a_ser_removido, filho_esq, filho_dir, pai)

  def caso_sem_filhos(self, vertice_a_ser_removido, pai):
      if pai is not None:

        if pai.get_filho_direita().get_chave() == vertice_a_ser_removido.get_chave():
          pai.set_filho_direita(None)

        else:
          pai.set_filho_esquerda(None)

      vertice_a_ser_removido = None

  def caso_filho_unico(self, vertice_a_ser_removido, filho_esq, filho_dir, pai):

    if filho_dir is not None:
      filho = filho_dir

    elif filho_esq is not None:
      filho = filho_esq

    else:
      raise KeyError("Ambos os filhos de um vértice com 1 filho são nulos.")

    filho.set_pai(pai)

    if pai is None:
      self.set_raiz(None)
      self.set_raiz(filho)

    else:

      if vertice_a_ser_removido.get_chave() == pai.get_filho_direita().get_chave():
        pai.set_filho_direita(filho)

      else:
        pai.set_filho_esquerda(filho)

  def detecta_sucessor(self, vertice):
    sucessor = vertice.get_filho_esquerda()
    filho_a_direita_do_possivel_sucessor = sucessor.get_filho_direita()

    while filho_a_direita_do_possivel_sucessor is not None:
      sucessor = filho_a_direita_do_possivel_sucessor
      filho_a_direita_do_possivel_sucessor = sucessor.get_filho_direita()

    return sucessor

  def caso_filho_duplo(self, vertice_a_ser_removido, filho_esq, filho_dir, pai):

    sucessor = self.detecta_sucessor(vertice_a_ser_removido)
    chave_sucessor = sucessor.get_chave()
    vertice_a_ser_removido.set_chave(chave_sucessor)

    if sucessor.get_filho_direita() is None and sucessor.get_filho_esquerda() is None:
      self.caso_sem_filhos(sucessor, sucessor.get_pai())

    else:
      self.caso_filho_unico(sucessor, sucessor.get_filho_esquerda(), sucessor.get_filho_direita(), sucessor.get_pai())

  def add_tamanho(self):
    self.__tamanho = self.__tamanho + 1

  def sub_tamanho(self):
    self.__tamanho = self.__tamanho - 1

  def percurso_pre_order(self, elementos_visitados = [], vertice = None):

    if vertice is None:
      vertice = self.get_raiz()
      elementos_visitados = []

    elementos_visitados.append(vertice.get_chave())

    if vertice.get_filho_esquerda() is not None:
      self.percurso_pre_order(elementos_visitados, vertice.get_filho_esquerda())

    if vertice.get_filho_direita() is not None:
      self.percurso_pre_order(elementos_visitados, vertice.get_filho_direita())

    return elementos_visitados

  def percurso_in_order(self, elementos_visitados = [], vertice = None):

    if vertice is None:
      vertice = self.get_raiz()
      elementos_visitados = []

    if vertice.get_filho_esquerda() is not None:
      self.percurso_in_order(elementos_visitados, vertice.get_filho_esquerda())

    elementos_visitados.append(vertice.get_chave())

    if vertice.get_filho_direita() is not None:
      self.percurso_in_order(elementos_visitados, vertice.get_filho_direita())

    return elementos_visitados

  def percurso_post_order(self, elementos_visitados = [], vertice = None):

    if vertice is None:
      vertice = self.get_raiz()
      elementos_visitados = []

    if vertice.get_filho_esquerda() is not None:
      self.percurso_post_order(elementos_visitados, vertice.get_filho_esquerda())

    if vertice.get_filho_direita() is not None:
      self.percurso_post_order(elementos_visitados, vertice.get_filho_direita())

    elementos_visitados.append(vertice.get_chave())

    return elementos_visitados

  def set_raiz(self, raiz):

    if self.__raiz is None or raiz is None:
      self.__raiz = raiz

    else:
      if raiz.get_chave() > self.__raiz.get_chave():
        raiz.set_filho_esquerda(self.__raiz)

      elif raiz.get_chave() < self.__raiz.get_chave():
        raiz.set_filho_direita(self.__raiz)

      self.__raiz.set_pai(raiz)

      self.__raiz = raiz

  def get_raiz(self):
    return self.__raiz

#------------------------------------------------------------------------------
```

### **Exercício 7**: Escreva um código que, dada uma árvore binária (não necessariamente de busca) na forma de um objeto do protótipo computacional em Programação Orientada a Objetos apresentado em aula e que a seus vértices sejam associados apenas valores numéricos, apresente o maior e o menor desses valores.

```python
# Solução:~>

class Vertice:
  # Atributos ------------------------------------------------------------------
  __pai               = None
  __filho_a_esquerda  = None
  __filho_a_direita   = None
  __chave             = 0
  # ----------------------------------------------------------------------------

  # Construtor -----------------------------------------------------------------
  def __init__(self, chave, pai = None, filho_esquerda = None, filho_direita = None):
    self.set_pai(pai)
    self.set_filho_esquerda(filho_esquerda)
    self.set_filho_direita(filho_direita)
    self.set_chave(chave)
  # ----------------------------------------------------------------------------

  # Métodos de configuração ----------------------------------------------------
  def set_pai(self, pai):
    self.__pai = pai

  def set_filho_esquerda(self, filho_esquerda):
    self.__filho_a_esquerda = filho_esquerda

  def set_filho_direita(self, filho_direita):
    self.__filho_a_direita = filho_direita

  def set_chave(self, chave):
    self.__chave = chave
  # ----------------------------------------------------------------------------

  # Métodos de retorno ---------------------------------------------------------
  def get_pai(self):
    return self.__pai

  def get_filho_esquerda(self):
    return self.__filho_a_esquerda

  def get_filho_direita(self):
    return self.__filho_a_direita

  def get_chave(self):
    return self.__chave
  # ----------------------------------------------------------------------------

class Arvore_Binaria_de_Busca:

  # Atributos ------------------------------------------------------------------
  __raiz    = None # Endereço da raíz da árvore
  __tamanho = 0
  __menor_numero_vertice = 0
  __maior_numero_vertice = 0
  # ----------------------------------------------------------------------------

  # Construtor -----------------------------------------------------------------
  def __init__(self, raiz):
    self.set_raiz(raiz)
    if raiz is not None:
      self.add_tamanho()
  # ----------------------------------------------------------------------------


  # Demais métodos -------------------------------------------------------------

  # ----------------------------------------------------------------------------
  # Inserção de vértices
  # ----------------------------------------------------------------------------
  def insere_vertice(self, novo_vertice):

      if self.get_raiz() is None:
        self.set_raiz(novo_vertice)

      else:
        self.aux_insere(novo_vertice, self.get_raiz())

  # ----------------------------------------------------------------------------
  def aux_insere(self, novo_vertice, possivel_pai):
    chave = novo_vertice.get_chave()

    if possivel_pai.get_chave() < chave:

      filho_dir = possivel_pai.get_filho_direita()

      if filho_dir is not None:
        self.aux_insere(novo_vertice, filho_dir)
      else:
        possivel_pai.set_filho_direita(novo_vertice)
        novo_vertice.set_pai(possivel_pai)

    elif possivel_pai.get_chave() > chave:

      filho_esq = possivel_pai.get_filho_esquerda()

      if filho_esq is not None:
        self.aux_insere(novo_vertice, filho_esq)
      else:
        possivel_pai.set_filho_esquerda(novo_vertice)
        novo_vertice.set_pai(possivel_pai)

    else:
      print("Chave já existente na árvore!")

  # ----------------------------------------------------------------------------
  # Busca
  # ----------------------------------------------------------------------------

  def busca(self, chave):

      temp = self.get_raiz()

      while temp.get_chave() != chave and temp is not None:

        chave_temporaria = temp.get_chave()

        if chave < chave_temporaria:
          temp = temp.get_filho_esquerda()
        elif chave > chave_temporaria:
          temp = temp.get_filho_direita()
        else:
          break


      if temp is None:
        print("Essa chave não existe na árvore!")
      else:
        return temp

  # ----------------------------------------------------------------------------
  # Remoção
  # ----------------------------------------------------------------------------
  # Analisa todos os casos para a remoção de um vértice: o caso do vértice ser
  # uma folha; o caso do vértice ter um filho único; e o caso do vértice ter
  # dois filhos.
  def remove_vertice(self, chave_a_ser_removida):

      temp = self.get_raiz()

      while temp.get_chave() != chave_a_ser_removida and temp is not None:

        chave_temporaria = temp.get_chave()

        if chave_a_ser_removida < chave_temporaria:
          temp = temp.get_filho_esquerda()
        elif chave_a_ser_removida > chave_temporaria:
          temp = temp.get_filho_direita()
        else:
          break


      if temp is None:
        print("Essa chave já não existe na árvore!")
      else:
        self.analisa_casos_de_remocao(temp)
        self.sub_tamanho()
  # ----------------------------------------------------------------------------

  # Uma vez que o vértice foi detectado, vamos considerar os casos para
  # removê-lo
  def analisa_casos_de_remocao(self, vertice_a_ser_removido):

      filho_esq = vertice_a_ser_removido.get_filho_esquerda()
      filho_dir = vertice_a_ser_removido.get_filho_direita()
      pai       = vertice_a_ser_removido.get_pai()

      # Se o vértice a ser removido for um vértice folha então basta anulá-lo.
      if filho_esq is None and filho_dir is None:
        self.caso_sem_filhos(vertice_a_ser_removido, pai)
      elif filho_esq is not None and filho_dir is not None:
        self.caso_filho_duplo(vertice_a_ser_removido, filho_esq, filho_dir, pai)
      else:
        self.caso_filho_unico(vertice_a_ser_removido, filho_esq, filho_dir, pai)
  # ----------------------------------------------------------------------------

  # Caso o vértice a ser removido seja um nó folha.
  def caso_sem_filhos(self, vertice_a_ser_removido, pai):
      if pai is not None:
        if pai.get_filho_direita().get_chave() == vertice_a_ser_removido.get_chave():
          pai.set_filho_direita(None)
        else:
          pai.set_filho_esquerda(None)

      vertice_a_ser_removido = None
  # ----------------------------------------------------------------------------

  # Caso o vértice a ser removido tenha apenas um filho
  def caso_filho_unico(self, vertice_a_ser_removido, filho_esq, filho_dir, pai):
    # O pai do vértice deve apontar para o filho do vértice.

    # Vamos descobrir onde está o filho do vértice (esquerda ou direita)
    if filho_dir is not None:
      filho = filho_dir
    elif filho_esq is not None:
      filho = filho_esq
    else:
      raise KeyError("Ambos os filhos de um vértice com 1 filho são nulos.")

    # Primeiramente, define-se o pai do vértice a ser removido como o novo pai
    # de seu filho.
    filho.set_pai(pai)

    # Se o vértice for a raíz, então ele não tem pai e, portanto, seu filho deve
    # se tornar a nova raíz.
    if pai is None:
      self.set_raiz(None)
      self.set_raiz(filho)
    else:

      # O vértice a ser removido é filho à direita ou a esquerda do pai?
      # Se for à direita, então seu filho deve ser apontado à direita pelo pai.
      if vertice_a_ser_removido.get_chave() == pai.get_filho_direita().get_chave():
        pai.set_filho_direita(filho)
      else:
      # Se for à esquerda, então seu filho deve ser apontado à esquerda pelo pai.
        pai.set_filho_esquerda(filho)
  # ----------------------------------------------------------------------------

  # Método para fazer a varredura à parte mais à direita da subárvore à esquerda
  # do ponto a ser removido. A este ponto dá-se o nome de sucessor do vértice
  # a ser removido.
  def detecta_sucessor(self, vertice):
    sucessor = vertice.get_filho_esquerda()
    filho_a_direita_do_possivel_sucessor = sucessor.get_filho_direita()

    while filho_a_direita_do_possivel_sucessor is not None:
      sucessor = filho_a_direita_do_possivel_sucessor
      filho_a_direita_do_possivel_sucessor = sucessor.get_filho_direita()

    return sucessor
  # ----------------------------------------------------------------------------

  # Método para deletar um vértice com dois filhos, sendo que, na verdade, neste
  # caso, não se faz uma remoção do vértice, mas sim sua substituição pelo seu
  # sucessor.
  def caso_filho_duplo(self, vertice_a_ser_removido, filho_esq, filho_dir, pai):

    # Primeiramente, detecta-se o sucesor do vértice.
    sucessor = self.detecta_sucessor(vertice_a_ser_removido)

    # Armazena o valor de sua chave.
    chave_sucessor = sucessor.get_chave()

    # Atualiza a chave do vértice a ser removido como sendo a chave do sucessor.
    vertice_a_ser_removido.set_chave(chave_sucessor)

    # Remove-se o sucessor do vértice da árvore.
    # Lembrando que o sucessor ou não tem filhos ou tem apenas 1 filho.
    if sucessor.get_filho_direita() is None and sucessor.get_filho_esquerda() is None:
      self.caso_sem_filhos(sucessor, sucessor.get_pai())
    else:
      self.caso_filho_unico(sucessor, sucessor.get_filho_esquerda(), sucessor.get_filho_direita(), sucessor.get_pai())
  # ----------------------------------------------------------------------------
  # ----------------------------------------------------------------------------


  # ----------------------------------------------------------------------------
  # Atualização do tamanho
  # ----------------------------------------------------------------------------
  def add_tamanho(self):
    self.__tamanho = self.__tamanho + 1

  def sub_tamanho(self):
    self.__tamanho = self.__tamanho - 1
  # ----------------------------------------------------------------------------

  def set_maior_vertice(self, vertice):
    self.__maior_numero_vertice = vertice

  def set_menor_vertice(self, vertice):
    self.__menor_numero_vertice = vertice

  def get_maior_vertice(self):
    return self.__maior_numero_vertice

  def get_menor_vertice(self):
    return self.__menor_numero_vertice

  def verificar_maior_menor_vertice(self, vertice):
    if self.get_menor_vertice() > vertice.get_chave():
      self.set_menor_vertice(vertice.get_chave())

    elif self.get_maior_vertice() < vertice.get_chave():
      self.set_maior_vertice(vertice.get_chave())

  def buscar_maior_menor_vertice(self, elementos_visitados = [], vertice = None):

    if vertice is None:
      vertice = self.get_raiz()
      self.set_maior_vertice(vertice.get_chave())
      self.set_menor_vertice(vertice.get_chave())
      elementos_visitados = []
    else:
      self.verificar_maior_menor_vertice(vertice)


    # visita
    elementos_visitados.append(vertice.get_chave())

    # vai para a árvore da esquerda
    if vertice.get_filho_esquerda() is not None:
      self.buscar_maior_menor_vertice(elementos_visitados, vertice.get_filho_esquerda())

    # vai para a árvore da direita
    if vertice.get_filho_direita() is not None:

      self.buscar_maior_menor_vertice(elementos_visitados, vertice.get_filho_direita())


    return self.get_menor_vertice(), self.get_maior_vertice()

  # ----------------------------------------------------------------------------
  # Percursos
  # ----------------------------------------------------------------------------

  def percurso_pre_order(self, elementos_visitados = [], vertice = None):

    if vertice is None:
      vertice = self.get_raiz()
      elementos_visitados = []

    # visita
    elementos_visitados.append(vertice.get_chave())

    # vai para a árvore da esquerda
    if vertice.get_filho_esquerda() is not None:
      self.percurso_pre_order(elementos_visitados, vertice.get_filho_esquerda())

    # vai para a árvore da direita
    if vertice.get_filho_direita() is not None:
      self.percurso_pre_order(elementos_visitados, vertice.get_filho_direita())

    return elementos_visitados





  # ----------------------------------------------------------------------------



  def percurso_in_order(self, elementos_visitados = [], vertice = None):

    if vertice is None:
      vertice = self.get_raiz()
      elementos_visitados = []

    # vai para a árvore da esquerda
    if vertice.get_filho_esquerda() is not None:
      self.percurso_in_order(elementos_visitados, vertice.get_filho_esquerda())

    # visita
    elementos_visitados.append(vertice.get_chave())

    # vai para a árvore da direita
    if vertice.get_filho_direita() is not None:
      self.percurso_in_order(elementos_visitados, vertice.get_filho_direita())

    return elementos_visitados
  # ----------------------------------------------------------------------------


  def percurso_post_order(self, elementos_visitados = [], vertice = None):

    if vertice is None:
      vertice = self.get_raiz()
      elementos_visitados = []

    # vai para a árvore da esquerda
    if vertice.get_filho_esquerda() is not None:
      self.percurso_post_order(elementos_visitados, vertice.get_filho_esquerda())

    # vai para a árvore da direita
    if vertice.get_filho_direita() is not None:
      self.percurso_post_order(elementos_visitados, vertice.get_filho_direita())

    # visita
    elementos_visitados.append(vertice.get_chave())

    return elementos_visitados
  # ----------------------------------------------------------------------------
  # ----------------------------------------------------------------------------



  # Configuração ---------------------------------------------------------------
  def set_raiz(self, raiz):
    if self.__raiz is None or raiz is None:
      self.__raiz = raiz
    else:
      if raiz.get_chave() > self.__raiz.get_chave():
        raiz.set_filho_esquerda(self.__raiz)
      elif raiz.get_chave() < self.__raiz.get_chave():
        raiz.set_filho_direita(self.__raiz)

      self.__raiz.set_pai(raiz)

      self.__raiz = raiz
  # ----------------------------------------------------------------------------


  # Retorno --------------------------------------------------------------------
  def get_raiz(self):
    return self.__raiz
  # ----------------------------------------------------------------------------

ABB = Arvore_Binaria_de_Busca(None)

vertice_raiz = Vertice(20)
ABB.insere_vertice(vertice_raiz)

vertice = Vertice(10)
ABB.insere_vertice(vertice)

vertice = Vertice(50)
ABB.insere_vertice(vertice)

vertice = Vertice(25)
ABB.insere_vertice(vertice)

vertice = Vertice(51)
ABB.insere_vertice(vertice)

vertice = Vertice(21)
ABB.insere_vertice(vertice)

vertice = Vertice(35)
ABB.insere_vertice(vertice)

vertice = Vertice(26)
ABB.insere_vertice(vertice)
print(40*'-')
print("Raíz da árvore: ", ABB.get_raiz().get_chave())
print(40*'-')

print("Filho à esquerda da árvore: ", ABB.get_raiz().get_filho_esquerda().get_chave())
print("Filho à direita da árvore: ", ABB.get_raiz().get_filho_direita().get_chave())
print(40*'-')
#ESQUERDA DE 10
print("Neto à esquerda do filho 10: ", ABB.get_raiz().get_filho_esquerda().get_filho_esquerda())
#DIREITA DE 10
print("Neto à direita do filho 10: ", ABB.get_raiz().get_filho_esquerda().get_filho_direita())
#ESQUERDA DE 50
print("Neto à esquerda do filho 50: ", ABB.get_raiz().get_filho_direita().get_filho_esquerda().get_chave())
#DIREITA DE 50
print("Neto à direita do filho 50: ", ABB.get_raiz().get_filho_direita().get_filho_direita().get_chave())
#ESQUERDA DE 51
print(40*'-')
print("Bisneto à esquerda do filho 51: ", ABB.get_raiz().get_filho_direita().get_filho_direita().get_filho_esquerda())
#DIREITA DE 51
print("Bisneto à direita do filho 51: ", ABB.get_raiz().get_filho_direita().get_filho_direita().get_filho_direita())
#ESQURDA DE 25
print("Bisneto à direita do filho 25: ", ABB.get_raiz().get_filho_direita().get_filho_esquerda().get_filho_esquerda().get_chave())
#DIREITA DE 25
print("Bisneto à direita do filho 25: ", ABB.get_raiz().get_filho_direita().get_filho_esquerda().get_filho_direita().get_chave())
#ESQUERDA DE 21
print(40*'-')
print("Tataraneto à direita do filho 21: ", ABB.get_raiz().get_filho_direita().get_filho_esquerda().get_filho_esquerda().get_filho_esquerda())
#DIREITA DE 21
print("Tataraneto à direita do filho 21: ", ABB.get_raiz().get_filho_direita().get_filho_esquerda().get_filho_esquerda().get_filho_direita())
#ESQUERDA DE 35
print("Tataraneto à direita do filho 35: ", ABB.get_raiz().get_filho_direita().get_filho_esquerda().get_filho_direita().get_filho_esquerda().get_chave())
#DIREITA DE 35
print("Tataraneto à direita do filho 35: ", ABB.get_raiz().get_filho_direita().get_filho_esquerda().get_filho_direita().get_filho_direita())
#ESQUERDA DE 26
print(40*'-')
print("Tatataraneto à direita do filho 26: ", ABB.get_raiz().get_filho_direita().get_filho_esquerda().get_filho_direita().get_filho_esquerda().get_filho_esquerda())
#DIREITA DE 26
print("Tatataraneto à direita do filho 26: ", ABB.get_raiz().get_filho_direita().get_filho_esquerda().get_filho_direita().get_filho_esquerda().get_filho_direita())
print(40*'-')
print("Percurso pre-order:", ABB.percurso_pre_order())
print(40*'-')

menor, maior = ABB.buscar_maior_menor_vertice()


print("Menor vertice:", menor,"\tMaior vertice:", maior)
```

### **Exercício 8**: Escreva um programa em Python que, dada uma árvore qualquer em notação de colchêtes aninhados, calcule e apresente a altura dessa árvore.

```python
# Solução:~>

class Vertice:
  # Atributos ------------------------------------------------------------------
  __pai               = None
  __filho_a_esquerda  = None
  __filho_a_direita   = None
  __chave             = 0
  # ----------------------------------------------------------------------------

  # Construtor -----------------------------------------------------------------
  def __init__(self, chave, pai = None, filho_esquerda = None, filho_direita = None):
    self.set_pai(pai)
    self.set_filho_esquerda(filho_esquerda)
    self.set_filho_direita(filho_direita)
    self.set_chave(chave)
  # ----------------------------------------------------------------------------

  # Métodos de configuração ----------------------------------------------------
  def set_pai(self, pai):
    self.__pai = pai

  def set_filho_esquerda(self, filho_esquerda):
    self.__filho_a_esquerda = filho_esquerda

  def set_filho_direita(self, filho_direita):
    self.__filho_a_direita = filho_direita

  def set_chave(self, chave):
    self.__chave = chave
  # ----------------------------------------------------------------------------

  # Métodos de retorno ---------------------------------------------------------
  def get_pai(self):
    return self.__pai

  def get_filho_esquerda(self):
    return self.__filho_a_esquerda

  def get_filho_direita(self):
    return self.__filho_a_direita

  def get_chave(self):
    return self.__chave
  # ----------------------------------------------------------------------------

class Arvore_Binaria_de_Busca:

  # Atributos ------------------------------------------------------------------
  __raiz    = None # Endereço da raíz da árvore
  __tamanho = 0
  __tamanho_vertice = 0

  # ----------------------------------------------------------------------------

  # Construtor -----------------------------------------------------------------
  def __init__(self, raiz):
    self.set_raiz(raiz)
    if raiz is not None:
      self.add_tamanho()
  # ----------------------------------------------------------------------------


  # Demais métodos -------------------------------------------------------------

  # ----------------------------------------------------------------------------
  # Inserção de vértices
  # ----------------------------------------------------------------------------

  def verifica_se_lista(self, possivel_lista):
    if isinstance(possivel_lista, list):
      for vertice in possivel_lista:
        self.verifica_se_lista(vertice)
    else:
      vertice = Vertice(possivel_lista)
      self.insere_vertice(vertice)

  def insere_vertice(self, novo_vertice):

      if isinstance(novo_vertice, list):
        self.verifica_se_lista(novo_vertice)

      else:

        if self.get_raiz() is None:
          self.set_raiz(novo_vertice)

        else:
          self.aux_insere(novo_vertice, self.get_raiz())

  # ----------------------------------------------------------------------------
  def aux_insere(self, novo_vertice, possivel_pai):
    chave = novo_vertice.get_chave()

    if possivel_pai.get_chave() < chave:

      filho_dir = possivel_pai.get_filho_direita()

      if filho_dir is not None:
        self.aux_insere(novo_vertice, filho_dir)
      else:
        possivel_pai.set_filho_direita(novo_vertice)
        novo_vertice.set_pai(possivel_pai)

    elif possivel_pai.get_chave() > chave:

      filho_esq = possivel_pai.get_filho_esquerda()

      if filho_esq is not None:
        self.aux_insere(novo_vertice, filho_esq)
      else:
        possivel_pai.set_filho_esquerda(novo_vertice)
        novo_vertice.set_pai(possivel_pai)

    else:
      print("Chave já existente na árvore!")

  # ----------------------------------------------------------------------------
  # Busca
  # ----------------------------------------------------------------------------

  def busca(self, chave):

      temp = self.get_raiz()

      while temp.get_chave() != chave and temp is not None:

        chave_temporaria = temp.get_chave()

        if chave < chave_temporaria:
          temp = temp.get_filho_esquerda()
        elif chave > chave_temporaria:
          temp = temp.get_filho_direita()
        else:
          break


      if temp is None:
        print("Essa chave não existe na árvore!")
      else:
        return temp

  # ----------------------------------------------------------------------------
  # Remoção
  # ----------------------------------------------------------------------------
  # Analisa todos os casos para a remoção de um vértice: o caso do vértice ser
  # uma folha; o caso do vértice ter um filho único; e o caso do vértice ter
  # dois filhos.
  def remove_vertice(self, chave_a_ser_removida):

      temp = self.get_raiz()

      while temp.get_chave() != chave_a_ser_removida and temp is not None:

        chave_temporaria = temp.get_chave()

        if chave_a_ser_removida < chave_temporaria:
          temp = temp.get_filho_esquerda()
        elif chave_a_ser_removida > chave_temporaria:
          temp = temp.get_filho_direita()
        else:
          break


      if temp is None:
        print("Essa chave já não existe na árvore!")
      else:
        self.analisa_casos_de_remocao(temp)
        self.sub_tamanho()
  # ----------------------------------------------------------------------------

  # Uma vez que o vértice foi detectado, vamos considerar os casos para
  # removê-lo
  def analisa_casos_de_remocao(self, vertice_a_ser_removido):

      filho_esq = vertice_a_ser_removido.get_filho_esquerda()
      filho_dir = vertice_a_ser_removido.get_filho_direita()
      pai       = vertice_a_ser_removido.get_pai()

      # Se o vértice a ser removido for um vértice folha então basta anulá-lo.
      if filho_esq is None and filho_dir is None:
        self.caso_sem_filhos(vertice_a_ser_removido, pai)
      elif filho_esq is not None and filho_dir is not None:
        self.caso_filho_duplo(vertice_a_ser_removido, filho_esq, filho_dir, pai)
      else:
        self.caso_filho_unico(vertice_a_ser_removido, filho_esq, filho_dir, pai)
  # ----------------------------------------------------------------------------

  # Caso o vértice a ser removido seja um nó folha.
  def caso_sem_filhos(self, vertice_a_ser_removido, pai):
      if pai is not None:
        if pai.get_filho_direita().get_chave() == vertice_a_ser_removido.get_chave():
          pai.set_filho_direita(None)
        else:
          pai.set_filho_esquerda(None)

      vertice_a_ser_removido = None
  # ----------------------------------------------------------------------------

  # Caso o vértice a ser removido tenha apenas um filho
  def caso_filho_unico(self, vertice_a_ser_removido, filho_esq, filho_dir, pai):
    # O pai do vértice deve apontar para o filho do vértice.

    # Vamos descobrir onde está o filho do vértice (esquerda ou direita)
    if filho_dir is not None:
      filho = filho_dir
    elif filho_esq is not None:
      filho = filho_esq
    else:
      raise KeyError("Ambos os filhos de um vértice com 1 filho são nulos.")

    # Primeiramente, define-se o pai do vértice a ser removido como o novo pai
    # de seu filho.
    filho.set_pai(pai)

    # Se o vértice for a raíz, então ele não tem pai e, portanto, seu filho deve
    # se tornar a nova raíz.
    if pai is None:
      self.set_raiz(None)
      self.set_raiz(filho)
    else:

      # O vértice a ser removido é filho à direita ou a esquerda do pai?
      # Se for à direita, então seu filho deve ser apontado à direita pelo pai.
      if vertice_a_ser_removido.get_chave() == pai.get_filho_direita().get_chave():
        pai.set_filho_direita(filho)
      else:
      # Se for à esquerda, então seu filho deve ser apontado à esquerda pelo pai.
        pai.set_filho_esquerda(filho)
  # ----------------------------------------------------------------------------

  # Método para fazer a varredura à parte mais à direita da subárvore à esquerda
  # do ponto a ser removido. A este ponto dá-se o nome de sucessor do vértice
  # a ser removido.
  def detecta_sucessor(self, vertice):
    sucessor = vertice.get_filho_esquerda()
    filho_a_direita_do_possivel_sucessor = sucessor.get_filho_direita()

    while filho_a_direita_do_possivel_sucessor is not None:
      sucessor = filho_a_direita_do_possivel_sucessor
      filho_a_direita_do_possivel_sucessor = sucessor.get_filho_direita()

    return sucessor
  # ----------------------------------------------------------------------------

  # Método para deletar um vértice com dois filhos, sendo que, na verdade, neste
  # caso, não se faz uma remoção do vértice, mas sim sua substituição pelo seu
  # sucessor.
  def caso_filho_duplo(self, vertice_a_ser_removido, filho_esq, filho_dir, pai):

    # Primeiramente, detecta-se o sucesor do vértice.
    sucessor = self.detecta_sucessor(vertice_a_ser_removido)

    # Armazena o valor de sua chave.
    chave_sucessor = sucessor.get_chave()

    # Atualiza a chave do vértice a ser removido como sendo a chave do sucessor.
    vertice_a_ser_removido.set_chave(chave_sucessor)

    # Remove-se o sucessor do vértice da árvore.
    # Lembrando que o sucessor ou não tem filhos ou tem apenas 1 filho.
    if sucessor.get_filho_direita() is None and sucessor.get_filho_esquerda() is None:
      self.caso_sem_filhos(sucessor, sucessor.get_pai())
    else:
      self.caso_filho_unico(sucessor, sucessor.get_filho_esquerda(), sucessor.get_filho_direita(), sucessor.get_pai())
  # ----------------------------------------------------------------------------
  # ----------------------------------------------------------------------------


  # ----------------------------------------------------------------------------
  # Atualização do tamanho
  # ----------------------------------------------------------------------------
  def add_tamanho(self):
    self.__tamanho = self.__tamanho + 1

  def sub_tamanho(self):
    self.__tamanho = self.__tamanho - 1
  # ----------------------------------------------------------------------------


  # ----------------------------------------------------------------------------
  # Percursos
  # ----------------------------------------------------------------------------

  def percurso_pre_order(self, elementos_visitados = [], vertice = None):

    if vertice is None:
      vertice = self.get_raiz()
      elementos_visitados = []

    # visita
    elementos_visitados.append(vertice.get_chave())

    # vai para a árvore da esquerda
    if vertice.get_filho_esquerda() is not None:
      self.percurso_pre_order(elementos_visitados, vertice.get_filho_esquerda())

    # vai para a árvore da direita
    if vertice.get_filho_direita() is not None:
      self.percurso_pre_order(elementos_visitados, vertice.get_filho_direita())

    return elementos_visitados

  # ----------------------------------------------------------------------------



  def percurso_in_order(self, elementos_visitados = [], vertice = None):

    if vertice is None:
      vertice = self.get_raiz()
      elementos_visitados = []

    # vai para a árvore da esquerda
    if vertice.get_filho_esquerda() is not None:
      self.percurso_in_order(elementos_visitados, vertice.get_filho_esquerda())

    # visita
    elementos_visitados.append(vertice.get_chave())

    # vai para a árvore da direita
    if vertice.get_filho_direita() is not None:
      self.percurso_in_order(elementos_visitados, vertice.get_filho_direita())

    return elementos_visitados
  # ----------------------------------------------------------------------------


  def percurso_post_order(self, elementos_visitados = [], vertice = None):

    if vertice is None:
      vertice = self.get_raiz()
      elementos_visitados = []

    # vai para a árvore da esquerda
    if vertice.get_filho_esquerda() is not None:
      self.percurso_post_order(elementos_visitados, vertice.get_filho_esquerda())

    # vai para a árvore da direita
    if vertice.get_filho_direita() is not None:
      self.percurso_post_order(elementos_visitados, vertice.get_filho_direita())

    # visita
    elementos_visitados.append(vertice.get_chave())

    return elementos_visitados
  # ----------------------------------------------------------------------------




  # Configuração ---------------------------------------------------------------
  def set_raiz(self, raiz):
    if self.__raiz is None or raiz is None:
      self.__raiz = raiz
    else:
      if raiz.get_chave() > self.__raiz.get_chave():
        raiz.set_filho_esquerda(self.__raiz)
      elif raiz.get_chave() < self.__raiz.get_chave():
        raiz.set_filho_direita(self.__raiz)

      self.__raiz.set_pai(raiz)

      self.__raiz = raiz
  # ----------------------------------------------------------------------------

  def set_tamanho_vertice(self, tamanho):
    self.__tamanho_vertice = tamanho

  def get_tamanho_vertice(self):
    return self.__tamanho_vertice

  # Retorno --------------------------------------------------------------------
  def get_raiz(self):
    return self.__raiz
  # ----------------------------------------------------------------------------


  # ----------------------------------------------------------------------------


  def calcular_tamanho_vertice(self, elementos_visitados = [], vertice = None, indice = 0):

    if self.get_tamanho_vertice() < indice:
      self.set_tamanho_vertice(indice)

    if vertice is None:
      vertice = self.get_raiz()
      indice += 1
      elementos_visitados = []

    # vai para a árvore da esquerda
    if vertice.get_filho_esquerda() is not None:
      indice += 1
      self.calcular_tamanho_vertice(elementos_visitados, vertice.get_filho_esquerda(), indice)
      indice -= 1


    # vai para a árvore da direita
    if vertice.get_filho_direita() is not None:
      indice += 1
      self.calcular_tamanho_vertice(elementos_visitados, vertice.get_filho_direita(), indice)
      indice -= 1


    # visita
    elementos_visitados.append(vertice.get_chave())

    return self.get_tamanho_vertice()


arvore = [20, [10], [50, [25, [21], [35, [26]]], [51]]]
avr = Arvore_Binaria_de_Busca(None)

avr.insere_vertice(arvore)

avr.calcular_tamanho_vertice()
```
