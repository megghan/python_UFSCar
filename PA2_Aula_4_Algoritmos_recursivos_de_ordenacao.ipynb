# Algoritmo ***Quick Sort*** e suas complexidades

Nesta técnica, o principal intuito consiste em posicionar um determinado elemento ```e``` de uma lista ```L``` de $n$ elementos em sua devida posição por meio do seguinte critério:

- Todos os elementos de ```L``` que sejam maiores do que ```e``` devem ser posicionados, respeitando-se sua disposição original, à direita de ```e```;
- Todos os elementos de ```L``` que sejam menores do que ```e``` devem ser posicionados, respeitando-se sua disposição original, à esquerda de ```e```;
- O mesmo procedimento deve ser efetuado sobre as sublistas geradas, respectivamente, pelos elementos à direita de ```e``` e pelos elementos à esquerda de ```e```;
- Se uma sublista é formada por apenas 1 elemento, então considera-se que a mesma já se encontra ordenada.


wuicksort.svg

Finalmente, o método de ordenação *Quick Sort* é definido através da seção de códigos a seguir:

````python
def quick_sort(L):

  n = len(L)

  if n <= 1:
    return L

  pivo = L[n-1]

  sublista_da_esquerda = []
  for i in range(n):
     if L[i] < pivo:
       sublista_da_esquerda.append(L[i])

  elementos_iguais_ao_pivo = []
  for i in range(n):
     if L[i] == pivo:
       elementos_iguais_ao_pivo.append(L[i])

  sublista_da_direita = []
  for i in range(n):
     if L[i] > pivo:
       sublista_da_direita.append(L[i])


  return quick_sort(sublista_da_esquerda) + elementos_iguais_ao_pivo + quick_sort(sublista_da_direita)

# Obs.: Lembre-se que o símbolo de '+' quando aplicado a listas possui o sentido
# de concatenação.
# Assim, [1,2,3] + [4,5,6,7,8] gera a lista [1,2,3,4,5,6,7,8].
````

Exemplo de funcionamento:

````python
lista_original = [2,3,2,4,2,2,5,43,5,6,7,5,655,65,65,756,757,5,765,57]
print("Lista original: ", lista_original)

lista_ordenada = quick_sort(lista_original)
print("Lista ordenada: ", lista_ordenada)
````

## **Análise de complexidade**

Considere o número de atribuições reais do algoritmo *Quick Sort*: $R_\text{Quick Sort}(n)$. Vamos analisar o funcionamento dessa função ao observar o algoritmo de ordenação em questão.

````python
def quick_sort(L):

  n = len(L) # 1 atribuição

  if n <= 1:
    return L

  pivo = L[n-1] # 1 atribuição

  sublista_da_esquerda = []
  for i in range(n):
     if L[i] < pivo:
       sublista_da_esquerda.append(L[i]) # n atribuições

  elementos_iguais_ao_pivo = []
  for i in range(n):
     if L[i] == pivo:
       elementos_iguais_ao_pivo.append(L[i]) # n atribuições

  sublista_da_direita = []
  for i in range(n):
     if L[i] > pivo:
       sublista_da_direita.append(L[i]) # n atribuições

  return quick_sort(sublista_da_esquerda) + elementos_iguais_ao_pivo + quick_sort(sublista_da_direita)
  #---------------------------------------------------------------------------------------------------
  # Número de atribuições efetuadas em quick_sort(sublista_da_esquerda)
  # +
  # Número de atribuições efetuadas em quick_sort(sublista_da_direita)
  #
  # Vamos supor que len(sublista_da_esquerda) = r
  # Vamos supor que len(sublista_da_direita)  = s
  #
  # R(n) = 2 + 3n + R(r) + R(s)
  #---------------------------------------------------------------------------------------------------
````

Supondo que a ordenação deve ser efetuada sobre os $n$ elementos de uma lista ```L```, repare que o número real de atribuições efetuadas no algoritmo *Quick Sort* é dado pela expressão:

<br>

$$R_\text{Quick Sort}(n) = 2 + 3n + R_\text{Quick Sort}(r) + R_\text{Quick Sort}(s),$$
<br>

sendo $r$ e $s$ os respectivos tamanhos das sublistas à esquerda e à direita do último elemento de ```L```.


Como $r$ e $s$ são valores menores que $n$, pois as sublistas geradas são estritamentes menores que a lista original, sabemos que $R_\text{Quick Sort}(r)$ e $R_\text{Quick Sort}(s)$ são estritamente menores que $R_\text{Quick Sort}(n)$. Além disso, quanto maiores $r$ e $s$ forem, maiores serão $R_\text{Quick Sort}(r)$ e $R_\text{Quick Sort}(s)$. Do mesmo modo, quanto menores $r$ e $s$, menores serão $R_\text{Quick Sort}(r)$ e $R_\text{Quick Sort}(s)$. Em outras palavras, quanto maior for uma sublista gerada em um nível da função, mais atribuições serão efetuadas. Ao mesmo tempo, quando menores forem ambas as sublistas geradas, menos atribuições serão efetuadas. Consequentemente, o pior caso deste algoritmo ocorre na configuração em que as sublistas são as mais desbalanceadas possíveis, enquanto que, o melhor caso deste algoritmo ocorre quando as listas são as mais balanceadas possíveis durante toda a execução.

Assim, para o pior caso, temos que apenas uma sublista (à esquerda ou à direita) será gerada para cada mergulho da função, enquanto que no melhor caso, as sublistas terão o mesmo tamanho que corresponde à metade da lista original.

No caso do código exemplo teríamos:

- Para o pior caso: $T_\text{Quick Sort}(n) = 2 + 3n + T_\text{Quick Sort}(n-1) + T_\text{Quick Sort}(0)$.

- Para o melhor caso: $B_\text{Quick Sort}(n) = 2 + 3n + B_\text{Quick Sort}\left(\frac{n}{2}\right) + B_\text{Quick Sort}\left(\frac{n}{2}\right)$.

Vamos analisar tais complexidades isoladamente.

## O **pior** caso do *Quick Sort*

Repare que $T_\text{Quick Sort}(0) = 0$, pois se a lista tem tamanho 0, o algoritmo já retorna a mesma lista, não necessitando conduzir nenhuma atribuição. Assim, podemos desenvolver a equação acima mencionada da seguinte maneira:


$$T_\text{Quick Sort}(n) = 2 + 3n + T_\text{Quick Sort}(n-1) + T_\text{Quick Sort}(0)$$

$$⇓$$

$$T_\text{Quick Sort}(n) = 2 + 3n + T_\text{Quick Sort}(n-1)$$
<br>
E, como a mesma regra sobre a lista e a geração de sublista será utilizada na análise de todos os níveis da recursão, podemos também obter as equações com respeito a $n-1, n-2, ..., 1$. Isto é,

<br>

$$T_\text{Quick Sort}(n-1) = 2 + 3(n-1) + T_\text{Quick Sort}(n-2)$$
$$T_\text{Quick Sort}(n-2) = 2 + 3(n-2) + T_\text{Quick Sort}(n-3)$$
$$\vdots$$
$$T_\text{Quick Sort}(2) = 2 + 3\cdot 2 + T_\text{Quick Sort}(1)$$
$$T_\text{Quick Sort}(1) = 2 + 3\cdot 1 + T_\text{Quick Sort}(0)$$

<br>

Somando todas as equações, teremos a equação final:

<br>

$$\sum_{i=1}^n T_\text{Quick Sort}(i) = \sum_{i=1}^n \left[ T_\text{Quick Sort}(i-1) + 3i + 2\right]$$

$$⇓$$


$$\sum_{i=1}^n T_\text{Quick Sort}(i) = \sum_{i=1}^n T_\text{Quick Sort}(i-1) + 3\sum_{i=1}^n i + \sum_{i=1}^n 2$$

$$⇓$$


$$\sum_{i=1}^n T_\text{Quick Sort}(i) = \sum_{i=1}^n T_\text{Quick Sort}(i-1) + 3n\frac{(n+1)}{2} + 2n$$

$$⇓$$


$$\sum_{i=1}^n\left[ T_\text{Quick Sort}(i) - T_\text{Quick Sort}(i-1)\right] = 3n\frac{(n+1)}{2} + 2n$$

$$⇓$$


$$ \left(T_\text{Quick Sort}(1) - T_\text{Quick Sort}(0)\right) + \left(T_\text{Quick Sort}(2) - T_\text{Quick Sort}(1)\right) \cdots + \left(T_\text{Quick Sort}(n-1) - T_\text{Quick Sort}(n-2)\right) + \left(T_\text{Quick Sort}(n) - T_\text{Quick Sort}(n-1)\right) = 3n\frac{(n+1)}{2} + 2n$$

$$⇓$$

$$T_\text{Quick Sort}(n) - T_\text{Quick Sort}(0) = 3n\frac{(n+1)}{2} + 2n$$

$$⇓   T_\text{Quick Sort}(0) = 0$$

$$T_\text{Quick Sort}(n) = 3n\frac{(n+1)}{2} + 2n$$

$$⇓$$

$$T_\text{Quick Sort}(n) = \frac{(3n^2+7n)}{2}$$

<br>

Logo, a complexidade para o pior caso é quadrática: $T_\text{Quick Sort}(n) = \mathcal{O}(n^2)$.

## O **melhor** caso do *Quick Sort*

Repare que $B_\text{Quick Sort}\left(1\right) = 0$, pois se a lista tem tamanho 1, o algoritmo já retorna a mesma lista, não necessitando conduzir nenhuma atribuição. Além disso, como estamos considerando o melhor caso, toda sublista gerada terá a metade do tamanho da sublista original. Sendo assim,

<br>
$$B_\text{Quick Sort}(n) = 2 + 3n + B_\text{Quick Sort}\left(\frac{n}{2}\right) + B_\text{Quick Sort}\left(\frac{n}{2}\right)$$

$$⇓$$

$$B_\text{Quick Sort}(n) = 2 + 3n + 2B_\text{Quick Sort}\left(\frac{n}{2}\right)$$

Além disso, focando a análise nas sublistas:


$$B_\text{Quick Sort}\left(\frac{n}{2}\right) = 2 + 3\frac{n}{2} + 2B_\text{Quick Sort}\left(\frac{n}{2^2}\right)$$

$$B_\text{Quick Sort}\left(\frac{n}{2^2}\right) = 2 + 3\frac{n}{2^2} + 2B_\text{Quick Sort}\left(\frac{n}{2^3}\right)$$

$$\vdots$$


$$B_\text{Quick Sort}\left(\frac{n}{2^k}\right) = 2 + 3\frac{n}{2^k} + 2B_\text{Quick Sort}\left(\frac{n}{2^{k+1}}\right)$$

$$\vdots$$

$$B_\text{Quick Sort}(2) = 2 + 3\cdot 2 + 2B_\text{Quick Sort}\left(1\right)$$


Substituindo $B_\text{Quick Sort}\left(\frac{n}{2}\right)$ na definição de $B_\text{Quick Sort}\left(n\right)$:

$$B_\text{Quick Sort}(n) = 2 + 3n + 2B_\text{Quick Sort}\left(\frac{n}{2}\right)$$

$$⇓$$

$$B_\text{Quick Sort}(n) = 2 + 3n + 2\left(2 + 3\frac{n}{2} + 2B_\text{Quick Sort}\left(\frac{n}{2^2}\right)\right)$$

$$⇓$$

$$B_\text{Quick Sort}(n) = 2 + 2^2 + 2\cdot(3n) + 2^2B_\text{Quick Sort}\left(\frac{n}{2^2}\right)$$


Substituindo $B_\text{Quick Sort}\left(\frac{n}{2^2}\right)$ na definição de $B_\text{Quick Sort}\left(n\right)$:


$$⇓$$

$$B_\text{Quick Sort}(n) = 2 + 2^2 + 2\cdot(3n) + 2^2\left(2 + 3\frac{n}{2^2} + 2B_\text{Quick Sort}\left(\frac{n}{2^3}\right)\right)$$


$$⇓$$

$$B_\text{Quick Sort}(n) = 2 + 2^2 + 2^3 + 3\cdot(3n) + 2^3B_\text{Quick Sort}\left(\frac{n}{2^3}\right)$$


$$⇓$$

$$\vdots$$


$$⇓$$

$$B_\text{Quick Sort}(n) = 2 + 2^2 + 2^3 + ... + 2^{k} + (k)\cdot(3n) + 2^{k}B_\text{Quick Sort}\left(\frac{n}{2^{k}}\right)$$


O maior valor possível para $k$ que podemos utilizar é aquele que reduz $\frac{n}{2^{k}}$ para $1$, que é $k = \log_2(n)$. Agregando esta informação às equações, teremos:

$$B_\text{Quick Sort}(n) = 2 + 2^2 + 2^3 + ... + 2^{\log_2(n)} + (\log_2(n))\cdot(3n) + 2^{\log_2(n)}B_\text{Quick Sort}\left(\frac{n}{2^{\log_2(n)}}\right)$$

$$⇓$$

$$B_\text{Quick Sort}(n) = 2 + 2^2 + 2^3 + ... + n + \log_2(n)\cdot(3n) + nB_\text{Quick Sort}\left(1\right)$$


Lembrando que $B_\text{Quick Sort}\left(1\right) = 0$, pois quando a lista possui tamanho 1, a função não faz nenhuma atribuição. Assim,


$$B_\text{Quick Sort}(n) = 2 + 2^2 + 2^3 + ... + n + 3n\log_2(n).$$


**Obs.**: $$ 2 + 2^2 + 2^3 + ... + n \leq \underbrace{n + n + n + ... + n}_{k = \log_2(n) \text{vezes}} = n \log_2(n).$$

Assim, na melhor situação, temos que a complexidade do método *Quick Sort* possui complexidade de ordem $n\text{log}_2(n)$. Isto é, o método é $\mathcal{O}(n\text{log}_2(n))$.

## O caso **médio** do *Quick Sort*:

Para o caso médio, vamos considerar que a cada mergulho da função de ordenação, a lista revese entre as leis que definem o melhor e o pior caso.


Prosseguindo dessa forma, vamos supor inicialmente que estamos diante do melhor caso. Assim, teremos que


$$M_\text{Quick Sort}(n) = B_\text{Quick Sort}(n) = 2 + 3n + 2M_\text{Quick Sort}\left(\frac{n}{2}\right).$$

Na sequência, vamos supor que ocorrerá o pior caso. Assim,


$$M_\text{Quick Sort}\left(\frac{n}{2}\right) = T_\text{Quick Sort}\left(\frac{n}{2}\right) =2 + 3\frac{n}{2} + M_\text{Quick Sort}\left(\frac{n}{2} - 1\right).$$

Substituindo esta informação na primeira equação, teremos:

$$M_\text{Quick Sort}(n)  = 2 + 3n + 2\left(2 + 3\frac{n}{2} + M_\text{Quick Sort}\left(\frac{n}{2} - 1\right)\right).$$

$$\Downarrow$$

$$M_\text{Quick Sort}(n) = 2 + 2^2 + 6n + 2M_\text{Quick Sort}\left(\frac{n}{2} - 1\right).$$

Repare que, se $n$ for um valor grande, podemos conduzir a  aproximação $\frac{n}{2}-1 ≈ \frac{n}{2}$, obtendo a seguinte relação:

$$M_\text{Quick Sort}(n) = 2 + 2^2 + 6n + 2M_\text{Quick Sort}\left(\frac{n}{2}\right).$$

Que é uma relação semelhante a que havíamos obtido para o melhor caso ($B_\text{Quick Sort}$), a qual possui desenvolvimento semelhante. Portanto, a ordem de complexidade para o caso médio do método de ordenação *Quick Sort* é dada por $\mathcal{O}(n\text{log}_2(n))$.


# Algoritmo ***Merge Sort*** e suas complexidades

A ideia principal do método ed ordenação ***Merge Sort*** consiste na estratégia de dividir para conquistar. Especificamente, divide-se a lista original em metades para que a rotina de ordenação ocorra em sublistas cada vez menores. Em teoria, a rotina de ordenação se inicia em sublistas de apenas 1 elemento, as quais já estão ordenadas por definição, e, na sequência, é feito a agregação dessas sublistas utilizando a estratégia conhecida como *merge*, ou mesclagem, que consiste em unir de forma ordenada duas sublistas já ordenadas, o que resulta em listas com dois elementos. Este processo de agregação é então conduzido nas sublistas geradas e, consequentemente, repetido até que uma lista do mesmo tamanho que a lista original seja gerada.

## A estratégia de mesclagem ordenada (***merge***):


Suponhamos que queremos unir duas listas ```lista_esquerda``` e ```lista_direita```, as quais já se encontram ordenadas, de forma a gerar uma lista ```lista_mesclada``` que também deve estar ordenada ao final de sua confecção.

Para isso, vamos conduzir a seguinte rotina:

- Vamos avaliar as duas listas ```lista_esquerda``` e ```lista_direita``` da esquerda para a direita;
- O menor elemento encontrado em cada lista nesse processo deve ser transferido para as posições desocupadas mais à esquerda da lista ```lista_mesclada``` resultante;
- Assim que uma das duas listas de entrada tiver todos os seus elementos transferidos para a lista resultante, os elementos restantes da outra lista devem ser transferidos para as últimas posições da lista mesclada.

Assim, é necessário conduzir uma rotina que controle as comparações entre os elementos das listas  ```lista_esquerda``` e ```lista_direita```, respeitando-se a orientação da esquerda para a direita. Ao mesmo tempo, esta rotina deve ir preenchendo as coordenadas da lista resultante.

merge.svg

A seguir, é apresentada a função no Python que executa a rotina de mesclagem ordenada.

````python
def merge(lista_esquerda, lista_direita):


  n_esquerda = len(lista_esquerda)
  n_direita  = len(lista_direita)

  #-----------------------------------------------------------------------------
  # Verificando os elementos de cada lista até que uma das duas acabe

  i_esq = 0
  i_dir = 0
  lista_mesclada = []
  while i_esq < n_esquerda and i_dir < n_direita:

      if lista_esquerda[i_esq] <= lista_direita[i_dir]:
          menor_valor = lista_esquerda[i_esq]
          i_esq = i_esq + 1
      else:
          menor_valor = lista_direita[i_dir]
          i_dir = i_dir + 1

      lista_mesclada.append(menor_valor)
  #-----------------------------------------------------------------------------


  #-----------------------------------------------------------------------------
  #-----------------------------------------------------------------------------
  #-----------------------------------------------------------------------------
  #  Preenchendo a lista mesclada com os valores restantes da lista ainda não
  # acabada
  #-----------------------------------------------------------------------------
  # Se i_esq == n_esquerda, este for não será executado
  for i in range(i_esq, n_esquerda):
      lista_mesclada.append(lista_esquerda[i])
  #-----------------------------------------------------------------------------

  #-----------------------------------------------------------------------------
  # Se i_dir == n_direita, este for não será executado
  for i in range(i_dir, n_direita):
      lista_mesclada.append(lista_direita[i])
  #-----------------------------------------------------------------------------
  #-----------------------------------------------------------------------------
  #-----------------------------------------------------------------------------

  return lista_mesclada
````

````python
lista_esquerda = [1,4,7,8]
lista_direita  = [2,3,5]

lista_mesclada = merge(lista_esquerda, lista_direita)

print("lista_mesclada = ", lista_mesclada)
````

## Uso de mesclagem em ordenação

O método ***Merge Sort*** é composto por duas etapas:

- **Dividir**: Dividir a lista em duas sublistas de mesmo tamanho (ou o mais próximo a isso) e, continuar essa divisão sobre as sublistas geradas até que não seja mais possível, isto é, até que a sublista gerada tenha apenas 1 elemento;
- **Conquistar**: Mesclar as sublistas geradas de forma ordenada (*merge*). Sendo que as listas menores possuem prioridade neste processo.

O processo é ilustrado no exemplo a seguir.

dividir.svg

conquistar.svg

A rotina em Python que descreve este processo é apresentada na sequência.

````python
def merge_sort(L):

  n = len(L)

  #  Por definição, uma lista vazia ou com apenas 1 elemento já se encontra
  # ordenada
  if n <= 1:
    return L

  #-----------------------------------------------------------------------------
  # Dividir a lista no meio
  ponto_medio = n // 2

  lista_esquerda = L[0:ponto_medio]
  lista_direita  = L[ponto_medio:n]
  #-----------------------------------------------------------------------------


  # Aplicar o algoritmo nas duas metades geradas -------------------------------
  lista_esquerda_ordenada = merge_sort(lista_esquerda)
  lista_direita_ordenada  = merge_sort(lista_direita)
  #-----------------------------------------------------------------------------

  # Agrupar as listas ordenadas com a técnica de mesclagem ---------------------
  L = merge(lista_esquerda_ordenada, lista_direita_ordenada)
  #-----------------------------------------------------------------------------

  return L
````

````python
L = [2,3,2,4,2,2,5,43,5,6,7,5,655,65,65,756,757,5,765,57]
print("Lista original: ", L)

Lista_ordenada = merge_sort(L)
print("Lista ordenada: ", Lista_ordenada)
````

## Análise de complexidade para o algoritmo ***Merge Sort***

Para analisarmos a complexidade do ***merge sort*** precisamos inicialmente analisar a complexidade da função ```merge```:


```python
def merge(lista_esquerda, lista_direita):


  n_esquerda = len(lista_esquerda) # 1 atribuição
  n_direita  = len(lista_direita)  # 1 atribuição
  
  i_esq = 0 # 1 atribuição
  i_dir = 0 # 1 atribuição
  lista_mesclada = [] # 1 atribuição
  while i_esq < n_esquerda and i_dir < n_direita:

      if lista_esquerda[i_esq] <= lista_direita[i_dir]:
          menor_valor = lista_esquerda[i_esq] # n_esquerda-1 execuções
          i_esq = i_esq + 1                   # n_esquerda-1 execuções
      else:
          menor_valor = lista_direita[i_dir]  # n_direita-1 execuções
          i_dir = i_dir + 1                   # n_direita-1 execuções

      lista_mesclada.append(menor_valor)      # n_direita + n_esquerda execuções
      
  # Não vão rodar---------------------------------------------------------------
  for i in range(i_esq, n_esquerda):
      lista_mesclada.append(lista_esquerda[i])
      
  for i in range(i_dir, n_direita):  
      lista_mesclada.append(lista_direita[i])
  #-----------------------------------------------------------------------------

  return lista_mesclada
```

Assim, a complexidade dessa função vai ser $T_\text{Merge}(n) = 5 + 3n$, sendo $n = n_\text{Esquerda} + n_\text{Direita}$ o tamanho da lista final gerada. Ou seja, a função de mesclagem ordenada possui complexidade de ordem $n$: $\mathcal{O}(n)$.

Repare também que o algoritmo de ordenação ***Merge Sort*** funcionará exatamente do mesmo modo independente da configuração da lista de entrada, uma vez que conduzirá o processo de repartição de listas sempre gerando duas sublistas de mesmo tamanho, o qual será aproximadamente a metade do tamanho da lista original. Assim, pode-se observar que uma lista ```L``` de entrada com tamanho $n$ será dividida em duas sublistas de tamanho aproximado $\frac{n}{2}$, as quais serão divididas em duas sublistas de tamanho aproximado $\frac{n/2}{2} = \frac{n}{2^2}$, e assim por diante até que as sublistas tenham tamanho $1$ ou $0$. Dessa forma, a função recursiva ```merge_sort``` terá conduzido $\log_2(n)$ mergulhos e, em cada um desses mergulhos, executará uma rotina (função ```merge```) de complexidade de ordem $n$, resultando em um algoritmo de complexidade de ordem $n\log_2(n)$ tanto no pior caso quanto no melhor caso. Sendo este um algoritmo de mesma complexidade em todos os casos.

Em detalhes, vamos modelar o número de atribuições efetuadas no algoritmo em questão:

```python
def merge_sort(L):

  n = len(L) # 1 atribuição

  if n <= 1:
    return L

  ponto_medio = n // 2 # 1 atribuição

  lista_esquerda = L[0:ponto_medio] # 1 atribuição
  lista_direita  = L[ponto_medio:n] # 1 atribuição
  
  lista_esquerda_ordenada = merge_sort(lista_esquerda) # T(n/2) atribuições
  lista_direita_ordenada  = merge_sort(lista_direita)  # T(n/2) atribuições


  L = merge(lista_esquerda_ordenada, lista_direita_ordenada) # 3n + 5 atribuições

  return L
```

Desse modo, o número de atribuições conduzidas no algoritmo se dá pelas seguintes equações:

$$T_\text{Merge Sort}(n) = 2T_\text{Merge Sort}\left(\frac{n}{2}\right) + 3n + 9 $$

$$T_\text{Merge Sort}\left(\frac{n}{2}\right) = 2T_\text{Merge Sort}\left(\frac{n}{4}\right) + 3\left(\frac{n}{2}\right) + 9 $$

$$T_\text{Merge Sort}\left(\frac{n}{4}\right) = 2T_\text{Merge Sort}\left(\frac{n}{8}\right) + 3\left(\frac{n}{4}\right) + 9 $$

$$\vdots$$

$$T_\text{Merge Sort}\left(2\right) = 2T_\text{Merge Sort}\left(1\right) + 3\left(2\right) + 9 $$

<br>

Substituindo a equação do nível abaixo na equação do nível acima, obteremos:

<br>

$$T_\text{Merge Sort}\left(n\right) = 2^kT_\text{Merge Sort}\left(\frac{n}{2^k}\right) + k\cdot 3n + 9 + 9\cdot 2 + 9 \cdot 2^2 + ... + 9\cdot 2^k. $$

<br>



Como o menor valor assumido nos argumentos de $T_\text{Merge Sort}$ é o valor $1$, então, o maior valor que $k$ pode assumir é tal que $\frac{n}{2^k} = 1$, ou seja, $k = \log_2(n)$. Substituindo este vamos na equação acima, obtemos:


<br>



$$T_\text{Merge Sort}\left(n\right) = 2^{\log_2(n)}T_\text{Merge Sort}\left(\frac{n}{2^{\log_2(n)}}\right) + \left(\log_2(n)\right)\cdot 3n + 9 + 9\cdot 2 + 9 \cdot 2^2 + ... + 9\cdot 2^{\log_2(n)}$$

$$⇓$$

$$T_\text{Merge Sort}\left(n\right) = nT_\text{Merge Sort}\left(1\right) + \left(\log_2(n)\right)\cdot 3n + 9 + 9\cdot 2 + 9 \cdot 2^2 + ... + 9\cdot n.$$


<br>


Como $T_\text{Merge Sort}\left(1\right) = 0$, pois se a lista possui apenas 1 elemento, então não há atribuições, então temos:


<br>


$$T_\text{Merge Sort}\left(n\right) = 3n\log_2(n) + 9 + 9\cdot 2 + 9 \cdot 2^2 + ... + 9\cdot n$$

$$⇓$$

$$T_\text{Merge Sort}\left(n\right) = 3n\log_2(n) + \mathcal{O}(n\log_2(n)).$$


<br>

Portanto, a ordem de complexidade do algoritmo ***Merge Sort*** é da ordem de $n\log_2(n)$. Além disso, como a complexidade do algoritmo é a mesma para todos os casos, então temos a seguinte equivalência:


<br>

$$T_\text{Merge Sort}\left(n\right) = B_\text{Merge Sort}\left(n\right) = M_\text{Merge Sort}\left(n\right) =  \mathcal{O}(n\log_2(n)).$$
